```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model= "openai/gpt-oss-20b",
)
```

```python
for chunk in llm.stream("Write me a 100 word paragraph on Artificial Intelligence."):
    print(chunk.text, end=" ", flush=True)
```

```python
for chunk in llm.stream("What is the anatomy of Cheetah which allows them to move at rapid speed?"):
    print(chunk.text, end="", flush=True)
```

```python
responses = llm.batch([
    "How do parrots speak?",
    "What is quantum mechanics?",
    "What is thermodynamics?"
])

for response in responses:
    print(response)
```

```python
llm.batch([
    "How do parrots speak?",
    "What is quantum mechanics?",
    "What is thermodynamics?"
    ],
    config = {
        'max_concurrency' : 5,
        }
)
```
